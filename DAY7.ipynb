DATA FILTERING &TRANSFORMATION

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Printer", "Monitor", "Keyboard"],
    "Sales": [50000, 30000, 20000, 15000, 25000, 10000],
    "Region": ["North", "South", "East", "West", "North", "South"]
}

df = pd.DataFrame(data)

print(df)
```

```python
south_sales=df[df["Region"]=="Sales"]
print(south_sales)
```

Sorting

```python
sorted_df = df.sort_values("Sales")
print(sorted_df)
```

Group by-Group similar records together

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

Aggregation

Sum()

```python
print(df["Sales"].sum())
```

mean()

```python
print(df["Sales"].mean())
```

Max()

```python
print(df["Sales"].max())
```

Min()

```python
print(df["Sales"].min())
```

Count()

```python
print(df["Sales"].count())
```

Std()

```python
print(df["Sales"].std())
```

Group calculation

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum", "mean", "max", "min"]
)

print(result)
```

Practice: Region-wise Sales Report

```python
import pandas as pd

data = {
    "Product": ["AC", "Refrigerator", "TV", "Watch", "AC", "TV"],
    "Sales": [45000, 35000, 25000, 10000, 50000, 30000]
}

df = pd.DataFrame(data)

result = df["Sales"].agg(["sum", "mean", "min", "max", "count"])

print(result)
```
