RESHAPE

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

new_arr = arr.reshape(2, 3)

print(new_arr)
```

TRANSPOSE

```python
import numpy as np

a = np.arange(1, 7)

b = a.reshape(2, 3)
print(b)

c = b.T
print(c)

d = a.reshape(-1, 2)
print(d)
```

Mean value

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

average = np.mean(arr)

print("Average =", average)
```

Median value

```python
import numpy as np

arr = np.array([10, 20, 30, 40,50])

print("Median =", np.median(arr))
```

Standard deviation

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Standard Deviation =", np.std(arr))
```

Employee Salary Analysis

```python
import numpy as np

employees = ["Ravi", "Priya", "Kumar", "Divya", "Arun"]

salary = np.array([25000, 30000, 28000, 35000, 32000])

# Total Salary
print("Total Salary:", np.sum(salary))

# Average Salary
print("Average Salary:", np.mean(salary))

# Highest Salary
print("Highest Salary:", np.max(salary))

# Lowest Salary
print("Lowest Salary:", np.min(salary))

# Highest Paid Employee
best_index = np.argmax(salary)
print("Highest Paid Employee:", employees[best_index])

# Employee Salary Report
print("\nEmployee Salary Report")
for i in range(len(employees)):
    print(employees[i], "-", salary[i])
```
