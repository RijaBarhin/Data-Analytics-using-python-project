ADVANCED VISUALIZATION

Histogram

```python
import matplotlib.pyplot as plt

expenses = [1200, 1500, 1800, 2000, 2200, 2500, 2800, 3000, 3200, 3500]

plt.hist(expenses,bins=12)

plt.title("Monthly Expense Distribution")
plt.xlabel("Expense Amount")
plt.ylabel("Frequency")

plt.show()
```

Scatter plots

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [35, 42, 50, 58, 65, 72, 80, 88]

plt.scatter(study_hours, marks)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

Sub plot

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)

plt.plot([5, 10, 15], [100, 200, 300])

plt.title("Revenue")

plt.subplot(1, 2, 2)

plt.plot([5, 10, 15], [300, 200, 100])

plt.title("Expense")

plt.show()
```
