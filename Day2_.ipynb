SIMPLE FUNCTION

Function Without Parameters

```python
def greet():
    print("Hello! Welcome.")

greet()
```

Function With Parameters

```python
def greet(name):
    print("Hello,", name)

greet("Rija")
```

Function With Return Value

```python
def add(a, b):
    return a + b

result = add(10, 20)
print("Sum =", result)
```
