DATA CLEANING WITH PANDAS

Missing Values

```python
import pandas as pd

data = {
    "Name": ["Rija", "Divya", "Kavi", "Arun"],
    "Marks": [85, None, 90, None]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.isnull())
```

dropna()

```python
clean_df = df.dropna()
print(clean_df)
```

fillna()

```python
df.fillna(0,inplace=True)
print(df)
```

Data Cleaning Using Mean Value

```python
import pandas as pd

data = {
    'Marks': [50, 60, 70, None]
}

df = pd.DataFrame(data)

mean_value = df['Marks'].mean()

df['Marks'].fillna(mean_value, inplace=True)

print(df)
```

Duplicates

```python
import pandas as pd

data = {
    "Name": ["Rija", "Divya", "Rija"],
    "Marks": [85, 90, 85]
}

df = pd.DataFrame(data)

print(df.duplicated())
```

remove duplicates

```python
import pandas as pd

data = {
    "Name": ["Riju", "Divya", "Riju"],
    "Marks": [85, 90, 85]
}

df = pd.DataFrame(data)
df =df.drop_duplicates()
print(df.duplicated())
```

Rename Columns

```python
import pandas as pd

data = {
    "Name": ["Riju", "Divya", "Arun"],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)

df = df.rename(columns={"Marks": "Score"})

print(df)
```

Employee Dataset Exploration

```python
import pandas as pd

employees = {
    "Emp_ID": [101, 102, 103, 104, 105],
    "Name": ["Arun", "Priya", "Rahul", "Divya", "Kumar"],
    "Department": ["IT", "HR", "IT", "Finance", "HR"],
    "Salary": [45000, 40000, 50000, 55000, 42000],
    "Experience": [2, 3, 5, 6, 4]
}

df = pd.DataFrame(employees)

print("Employee Dataset")
print(df)

print("\nAverage Salary")
print(df["Salary"].mean())

print("\nHighest Salary")
print(df["Salary"].max())

print("\nEmployees in HR")
print(df[df["Department"] == "HR"])

print("\nEmployees with Experience > 3")
print(df[df["Experience"] > 3])

print("\nEmployee Count by Department")
print(df["Department"].value_counts())
```

Clean a messy sales datasets

```python
import pandas as pd

data = {
    "Name": ["Riju", "Divya", "Divya", "Arun"],
    "Sales": [1000, None, None, 1500]
}

df = pd.DataFrame(data)

print("Original Data")
print(df)

print("\nMissing Values")
print(df.isnull())

df["Sales"] = df["Sales"].fillna(df["Sales"].mean())

df.drop_duplicates(inplace=True)

df.rename(columns={"Sales": "Total_Sales"}, inplace=True)

print("\nCleaned Data")
print(df)
```
