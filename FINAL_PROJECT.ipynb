HR EMPLOYEE ATTRITION ANALYSIS

Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading CSV File

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("HR-Employee-Attrition&Analysis .csv")

print(df.head())
```

Display Dataset Information

```python
print(df.info())
```

Display First 10 Records

```python
print(df.head(10))
```

Display Last 10 Records

```python
print(df.tail(10))
```

Find Missing Values

```python
print(df.isnull().sum())
```

Remove Missing Values - dropna()

```python
clean_df = df.dropna()

print(clean_df)
```

Fill Missing Values - fillna()

```python
filled_df = df.fillna(0)

print(filled_df)
```

Remove Duplicate Records

```python
print(df.duplicated().sum())

df = df.drop_duplicates()

print(df.shape)
```

Rename Columns

```python
df.rename(columns={'MonthlyIncome':'Salary'}, inplace=True)

print(df.columns)
```

Statistical Analysis - Mean

```python
print(df['Salary'].mean())
```

Statistical Analysis - Median

```python
print(df['Salary'].median())
```

Statistical Analysis - Standard Deviation

```python
print(df['Salary'].std())
```

DATA FILTERING


Employees with Salary greater than 5000

```python
high_salary = df[df['Salary'] > 5000]

print(high_salary)
```

SORTING DATA


Sort Employees by Salary

```python
sorted_data = df.sort_values(by='Salary', ascending=False)

print(sorted_data.head())
```

GROUP BY DEPARTMENT

```python
department_group = df.groupby('Department')['Salary'].mean()

print(department_group)
```

Aggregation - Sum

```python
print(df.groupby('Department')['Salary'].sum())
```

Aggregation - Mean

```python
print(df.groupby('Department')['Salary'].mean())
```

Aggregation - Maximum

```python
print(df.groupby('Department')['Salary'].max())
```

Aggregation - Minimum

```python
print(df.groupby('Department')['Salary'].min())
```

Aggregation - Count

```python
print(df.groupby('Department')['Salary'].count())
```

Aggregation - Standard Deviation

```python
print(df.groupby('Department')['Salary'].std())
```

DATA VISUALIZATION

Line Chart - Employee Age

```python
plt.plot(df['Age'].head(20))

plt.title("Employee Age Analysis")
plt.xlabel("Employees")
plt.ylabel("Age")

plt.show()
```

Bar Chart - Attrition Count

```python
df['Attrition'].value_counts().plot(kind='bar')

plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")

plt.show()
```

Pie Chart - Department Distribution

```python
df['Department'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Department Distribution")

plt.show()
```

Histogram - Age Distribution

```python
plt.hist(df['Age'])

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Scatter Plot - Age vs Salary

```python
plt.scatter(df['Age'], df['Salary'])

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```

Dashboard

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))

df['Age'].hist(ax=ax[0,0])
ax[0,0].set_title("Age Distribution")

df['Salary'].hist(ax=ax[0,1])
ax[0,1].set_title("Salary Distribution")

df['Department'].value_counts().plot(
    kind='bar',
    ax=ax[1,0]
)
ax[1,0].set_title("Department Count")=

df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=ax[1,1]
)
ax[1,1].set_title("Attrition Status")

plt.tight_layout()
plt.show()
```
