Student Score Program:

```python
scores = [78, 85, 92, 67, 88]

total = sum(scores)
average = total / len(scores)
highest = max(scores)
lowest = min(scores)

print("Student Scores:", scores)
print("Total Marks:", total)
print("Average Marks:", average)
print("Highest Mark:", highest)
print("Lowest Mark:", lowest)
```

Sales Array Calculations

```python
sales = [1200, 1500, 1800, 1000, 2000]

total_sales = sum(sales)
average_sales = total_sales / len(sales)
highest_sales = max(sales)
lowest_sales = min(sales)

print("Sales:", sales)
print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Highest Sales:", highest_sales)
print("Lowest Sales:", lowest_sales)
```

Attendance calculator

```python
def calculate_attendance(total, attended):
    percentage = (attended / total) * 100
    return percentage

try:
    total_classes = int(input("Enter Total Classes: "))
    attended_classes = int(input("Enter Attended Classes: "))

    attendance = calculate_attendance(total_classes, attended_classes)

    print("Attendance Percentage =", attendance, "%")

except ValueError:
    print("Invalid Input! Enter numbers only.")

except ZeroDivisionError:
    print("Total Classes cannot be zero.")
```

Average calculator

```python
def calculate_average(a, b, c):
    return (a + b + c) / 3

try:
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    num3 = float(input("Enter Third Number: "))

    avg = calculate_average(num1, num2, num3)

    print("Average =", avg)

except ValueError:
    print("Invalid Input! Enter numbers only.")
```
