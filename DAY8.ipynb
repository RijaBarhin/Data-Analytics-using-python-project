DATA VISUALIZATION BASICS

Line Chart

```python
import matplotlib.pyplot as plt

months = ["Jun", "Jul", "Aug", "Sep", "Oct"]
sales = [22000, 28000, 25000, 32000, 40000]

plt.plot(months, sales)

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jun", "Jul", "Aug", "Sep", "Oct"]

sales = [15000, 18000, 17000, 22000, 25000]

plt.plot(months, sales,
         color="red",
         marker="s",
         linestyle="-.")

plt.title("Monthly Sales Report")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt

students = ["K", "L", "M", "N"]

marks = [65, 88, 72, 96]

plt.bar(students, marks, color="purple")

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

Pie chart

```python
import matplotlib.pyplot as plt

sales = [4500, 2500, 3500]

products = ["Laptop", "Mobile", "Tablet"]

plt.pie(sales,
        labels=products)

plt.show()
```

Percentage identify &color change

```python
import matplotlib.pyplot as plt

sales = [4500, 2500, 3500]

products = ["Laptop", "Mobile", "Tablet"]

plt.pie(sales,
        labels=products,
        autopct='%1.1f%%',colors=["Red","blue","green"])

plt.show()
```

Monthly Sales dashboard

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [12000, 15000, 18000, 14000, 20000, 22000]

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.bar(months, sales)
plt.title("Bar Chart")

plt.subplot(1,3,2)
plt.plot(months, sales, marker='o')
plt.title("Line Chart")

plt.subplot(1,3,3)
plt.pie(sales, labels=months, autopct='%1.1f%%')
plt.title("Pie Chart")

plt.suptitle("Monthly Sales Dashboard")
plt.tight_layout()

plt.show()
```
