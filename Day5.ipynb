 Employee Dataset Exploration

```python
import pandas as pd

students = {
    "Roll_No": [1, 2, 3, 4, 5],
    "Name": ["Rija", "Divya", "Kavi", "Arun", "Priya"],
    "Department": ["BCA", "BCA", "BSc", "BCA", "BSc"],
    "Marks": [85, 90, 78, 88, 92],
    "Age": [19, 20, 19, 21, 20]
}

df = pd.DataFrame(students)

print("Student Dataset")
print(df)

print("\nAverage Marks")
print(df["Marks"].mean())

print("\nHighest Marks")
print(df["Marks"].max())

print("\nBCA Students")
print(df[df["Department"] == "BCA"])

print("\nStudents with Marks > 80")
print(df[df["Marks"] > 80])

print("\nStudent Count by Department")
print(df["Department"].value_counts())
```
